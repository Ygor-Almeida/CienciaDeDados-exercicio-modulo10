```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('ecommerce_preparados.csv')
```

```python
plt.hexbin(df['Preço'], df['Desconto'], gridsize=25, cmap='Blues')
plt.colorbar(label='Contagem da Concentração')
plt.xlabel('Preço')
plt.ylabel('Desconto')
plt.title('Disperção de Preço e Desconto')
plt.show()
```

```python
df.loc[df['Gênero'] == 'Meninas', 'Gênero'] = 'Feminino'
df.loc[df['Gênero'] == 'Mulher', 'Gênero'] = 'Feminino'
df.loc[df['Gênero'] == 'short menina verao look mulher', 'Gênero'] = 'Feminino'
df.loc[df['Gênero'] == 'roupa para gordinha pluss P ao 52', 'Gênero'] = 'Feminino'
df.loc[df['Gênero'] == 'bermuda feminina brilho Blogueira', 'Gênero'] = 'Feminino'
df.loc[df['Gênero'] == 'Bebês', 'Gênero'] = 'Sem gênero'
df.loc[df['Gênero'] == 'Unissex', 'Gênero'] = 'Sem gênero'
df.loc[df['Gênero'] == 'Sem gênero infantil', 'Gênero'] = 'Sem gênero'
df.loc[df['Gênero'] == 'menino', 'Gênero'] = 'Masculino'
df.loc[df['Gênero'] == 'Meninos', 'Gênero'] = 'Masculino'

x = df['Gênero'].value_counts().index
y = df['Gênero'].value_counts().values

plt.figure(figsize=(5, 5))
plt.pie(y, labels=x, autopct='%.1f%%', startangle=90)
plt.title('Distribuição de Gênero')
plt.show()
```

```python
x = df['Temporada_Cod'].value_counts().index
y = df['Temporada_Cod'].value_counts().values

plt.figure(figsize=(5, 5))
plt.bar(x, y, color='green')
plt.xlabel('Título')
plt.ylabel('Qtd_Vendidos')
plt.title('Vendas por Temporada')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.kdeplot(df['Preço'], fill=True, color='Red')
plt.xlabel('Preço')
plt.title('Densidade de Preço')
plt.show()
```

```python
df_corr = df[['Marca_Cod', 'Material_Cod', 'Nota']].corr()
plt.figure(figsize=(10, 5))
sns.heatmap(df_corr, annot=True, fmt='.2f')
plt.title('Correlação entre Marca, Material e Nota')
plt.show()
```
